# 9. Iterrows

```python
#iteracao linha a linha
```

```python
#Qual porto embarcou?  (C = Cherbourg; Q = Queenstown; S = Southampton)#
df['Embarked'].value_counts()
```

```python
df_20 = df[0:20]
```

```python
df_20
```

```python
for k, v in df_20.iterrows() : 
    print(k)
```

```python
for k, v in df_20.iterrows() : 
    print(v)
```

```python
Cherbourg = []

for index, row in df_20.iterrows() : 
    if row['Embarked'] == 'C': 
        Cherbourg.append('Yes')
    else: 
        Cherbourg.append('No')
```

```python
Cherbourg
```

```python
df_20['Cherbourg'] = Cherbourg
df_20
```
